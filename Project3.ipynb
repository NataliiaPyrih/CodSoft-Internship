##Machine Learning

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, precision_score, accuracy_score, recall_score, classification_report, confusion_matrix
```

```python
label=LabelEncoder()
df['diagnosis']=label.fit_transform(df['diagnosis'])
df.head()
```

```python
fig, ax = plt.subplots(figsize=(22, 12))
sns.heatmap(df.corr(), annot=True, linewidths=1, fmt='.2f')
```

```python
df.corr()['diagnosis'][:].sort_values(ascending=False)
```

```python
X = df.drop(['diagnosis', 'id'], axis=1)
y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=121, test_size=.3)
pd.DataFrame([y_train.value_counts(), y_test.value_counts()], index=['y_train', 'y_test'])
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
models={
        "LogisticRegression": LogisticRegression(),
        "KNeighborsClassifier": KNeighborsClassifier(),
        "DecisionTreeClassifier":DecisionTreeClassifier(),
        "RandomForestClassifier": RandomForestClassifier()
}
for model in models.values():
  model.fit(X_train_scaled, y_train)
```

```python
models_test_pred = {}

def evaluate_model(model, name, X_test, y_test):
    y_pred = model.predict(X_test_scaled)
    models_test_pred[name] = y_pred
    f1Score=f1_score(y_test, y_pred)
    precisionScore=precision_score(y_test, y_pred)
    accuracyScore=accuracy_score(y_test, y_pred)
    recallScore=recall_score(y_test, y_pred)
    return f1Score, precisionScore, accuracyScore, recallScore

f1_scores={}
precision_scores={}
accuracy_scores={}
recall_scores={}

for name, model in models.items():
  f1Score, precisionScore, accuracyScore, recallScore = evaluate_model(model, name, X_test_scaled, y_test)
  f1_scores[name]=f1Score
  precision_scores[name]=precisionScore
  accuracy_scores[name]=accuracyScore
  recall_scores[name]=recallScore

  print(f'{name} f1: {f1Score}')
  print(f'{name} precision: {precisionScore}')
  print(f'{name} accuracy: {accuracyScore}')
  print(f'{name} recall: {recallScore}\n')
```

```python
print(classification_report(y_test, models_test_pred['LogisticRegression']))
```

```python
sns.heatmap(confusion_matrix(y_test, models_test_pred['LogisticRegression']), annot=True, fmt='1')
```

```python
print(classification_report(y_test, models_test_pred['KNeighborsClassifier']))
```

```python
sns.heatmap(confusion_matrix(y_test, models_test_pred['KNeighborsClassifier']), annot=True, fmt='1')
```

```python
print(classification_report(y_test, models_test_pred['DecisionTreeClassifier']))
```

```python
sns.heatmap(confusion_matrix(y_test, models_test_pred['DecisionTreeClassifier']), annot=True, fmt='1')
```

```python
print(classification_report(y_test, models_test_pred['RandomForestClassifier']))
```

```python
sns.heatmap(confusion_matrix(y_test, models_test_pred['RandomForestClassifier']), annot=True, fmt='1')
```
